# The problem

A trading company is looking for a way to maximize profit per transportation of their goods. The company has a train available with 3 wagons.
When stocking the wagons they can choose among 4 types of cargo, each with its own specifications. How much of each cargo type should be loaded on which wagon in order to maximize profit?

## Dissection of the Problem
Before we do anything else, we need to consider what problem we’re trying to solve here exactly.

## The decision variables
The variables in question are the number of tonnes of each cargo type to load on each wagon, giving us a total of 12 variables to track.
<center>
    x<sub>ij</sub> = tonnes of cargo type i on wagon j
</center>
Notice that we are supposing here that each tonne can be fractioned (continuity) and each fraction allocated to a different wagon.

## The objective function
The objective function is the function that we seek to maximize or minimize. In this case, we want to maximize the profit, P, which is in turn a function of the twelve defined variables:
<center>
    P = <font face="Symbol">&#229;</font><sub>i</sub><font face="Symbol">&#229;</font><sub>j</sub>p<sub>i</sub>x<sub>ij</sub>
</center>

## The constraints
We summarize the constraints mentioned in the problem statements here:
<ul>
    <li>The total weight of the cargoes must not surpass wagons' limited weight capacities:
    <center>
    <font face="Symbol">&#229;</font><sub>i</sub>x<sub>ij</sub> <font face="Symbol">&#163;</font> w<sub>j</sub> <font face="Symbol">&#34;</font>j
    </center>
    <li>The total volume of the cargoes must not surpass wagons' limited volume capacities:
    <center>
    <font face="Symbol">&#229;</font><sub>i</sub>v<sub>i</sub>x<sub>ij</sub> <font face="Symbol">&#163;</font> s<sub>j</sub> <font face="Symbol">&#34;</font>j
    </center>
    <li>The usage of each cargo must not surpass its limited availability:
    <center>
    <font face="Symbol">&#229;</font><sub>i</sub>x<sub>ij</sub> <font face="Symbol">&#163;</font> a<sub>i</sub> <font face="Symbol">&#34;</font>i
    </center>
    <li>Clearly, the tonnes of cargo cannot be negative values:
    <center>
    x<sub>ij</sub> <font face="Symbol">&#179;</font> 0 <font face="Symbol">&#34;</font>i, <font face="Symbol">&#34;</font>j
    </center>
</ul>
So, now that we know all the relevant functions, we can now implement this using R and solving it by means of a linear programming solver.
